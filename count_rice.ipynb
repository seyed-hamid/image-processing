**قسمت اول** 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/Im491.jpg",cv2.IMREAD_GRAYSCALE)
#Binary
ret,binary = cv2.threshold(img,160,255,cv2.THRESH_BINARY)
plt.imshow(binary)
plt.title('binary img491')
plt.show()
#Gaussian Filter
dst = cv2.GaussianBlur(binary, (3,3), 0)
plt.imshow(dst)
plt.title("Gaussian filter img491")
plt.show()
#structuting element
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
plt.imshow(kernel)
plt.title('structuting element(5,5)')
plt.show()
#erosion
erosion = cv2.erode(dst,kernel)
plt.imshow(erosion)
plt.title("erosion img491")
plt.show()
# rice number
contours,hierarchy = cv2.findContours(erosion, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, contours, -1, (255, 0, 0), 2)
cv2_imshow(rgb)
print ("rice number= ",len(contours))
```
